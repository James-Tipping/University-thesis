```python
import uproot as up
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')
```

```python
# Opens background and signal files in dictionary format with variables
# The .keys function displays the headers of the dictionary
bkgd = up.open('ttbar.root')
btree = bkgd[b'ntuple;3']
btree.keys()
btree = btree.arrays()

sig = up.open('signal1.root')
sigtree = sig[b'EwkOneLeptonTwoBjets2018__ntuple;3']
print(sigtree.keys())
stree = sigtree.arrays()

hf = btree[b'HFScale']
print(hf)
```

```python
# Channels of background and signal plotted
bchannel = btree[b'mcChannel']
schannel = stree[b'mcChannel']

plt.hist(bchannel, label='Background', range=(392000, 420000), bins=100, density=True, histtype=u'step')
plt.hist(schannel, label='Signal', range=(392000, 420000), bins=100, density=True, histtype=u'step')
plt.xlabel('Channel')
plt.ylabel('Arbitrary units')
plt.title('Channels used')
plt.legend()
plt.show()
```
