```python
import uproot as up
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')
import configparser
```

```python
# Function to find the centre of the bins (given the edges) to plot bar charts
def centre(edges):
    a = np.array([])
    for i in range(len(edges)-1):
         a = np.append(a, ((edges[i]+edges[i+1])/2))
    return a

```

```python
def func(x, y):
    file = up.open(x)
    file = file['dataset;1']
    bdt = file['Method_BDT']['BDT']
    bdtvar = bdt['MVA_BDT_rejBvsS;1']

    keras = file['Method_PyKeras']['PyKeras']
    kerasvar = keras['MVA_PyKeras_rejBvsS']


    plt.plot(centre(kerasvar.edges), kerasvar.values, label='Neural Network')
    plt.plot(centre(bdtvar.edges), bdtvar.values, label='Boosted decision tree')
    plt.xlabel('Signal efficiency')
    plt.ylabel('Background rejection')
    plt.title('ROC curve')
    plt.legend()
    plt.show()

    config = configparser.ConfigParser()
    config.read(y)
    print('var= {}'.format(config.get('VARIABLES', 'training_variables')))
    print('gaussian transform {}'.format(config.get('VARIABLES', 'gaussian_transform')))
    print('BDT n trees {}'.format(config.get('HYPERPARAMETERS', 'BDT_n_trees')))
    print('DNN n layers {}'.format(config.get('HYPERPARAMETERS', 'DNN_n_layers')))
    print('DNN nodes hidden {}'.format(config.get('HYPERPARAMETERS', 'DNN_nodes_hidden')))
    print('DNN early stopping {}'.format(config.get('HYPERPARAMETERS', 'DNN_early_stopping')))
    print('DNN LR {}'.format(config.get('HYPERPARAMETERS', 'DNN_LR')))

    return

```
